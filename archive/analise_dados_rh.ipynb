```python
#Analise exploratoria de dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


dados = pd.read_csv("aug_train.csv")
#dados.head()
dados.info()



```

```python
dados.describe(include = object)
```

```python
dados.describe().drop(columns=['enrollee_id', 'target'])
```

```python
lista_colunas =  list(dados.columns.values[3:12])
```

```python
plt.figure(figsize=(18, 30))

a = 0
for i in lista_colunas:
    a += 1
    plt.subplot(3, 3, a)
    
    # Gráfico atualizado
    ax = sns.countplot(
        data=dados.fillna("NaN"), 
        x=i,
        hue=i,  # Adicionado para usar a paleta corretamente
        palette="viridis",
        legend=False,  # Remove a legenda redundante
        saturation=0.75
    )
    
    plt.title(i, fontsize=20)
    
    # Anotações (mantido igual)
    for p in ax.patches:
        ax.annotate(
            f"\n{p.get_height()}", 
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', 
            xytext=(0, 10), 
            textcoords='offset points',
            color='black'
        )
    
    if a > 7:
        plt.xticks(rotation=45)

plt.tight_layout(h_pad=2)
```
